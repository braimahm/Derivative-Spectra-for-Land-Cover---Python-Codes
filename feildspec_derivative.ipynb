```python
'''we plot the interpolated results against the created 10nm intervals
'''
plt.plot(xvals,y_interp)

plt.title("10nm Interpolated Reflectance")
plt.xlabel("Wavelength")
plt.ylabel("Reflectance")
plt.show()
```

```python
''' we want to claculate the derivative using the gradient function
dydx = np.gradient(y,x)
'''
derivative = np.gradient(y_interp, xvals)
plt.plot(xvals,derivative)
plt.title("Derivative Spectra on the 10nm Centres")
plt.xlabel("Wavelength")
plt.ylabel("Derivative Reflectance")

```

```python
''' We want to generate summary plot for the 1 and 10 nm,
 and the derivative spectra

'''
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(xvals, y_interp, 'ro')
ax1.plot(sk_1_grass.Wavelength, sk_1_grass.Mean, 'b-')
ax2.plot(xvals,derivative, "-")

ax1.set_xlabel("Wavelength")
ax1.set_ylabel("Reflectance")
ax2.set_ylabel("Derivative Spectra")
plt.title("Summary Spectral Plots")
ax1.legend(labels = ['10nm Reflectance','1nm Reflectance', 'Derivative Spectra'])
ax2.legend(labels = ["Derivative Spectra"], loc="upper left")
plt.show()


# plt.plot(xvals,y_interp, "*")
# plt.plot(sk_1_grass.Wavelength, sk_1_grass.Mean, "-")
# plt.plot(xvals,derivative, "--")
# plt.title("Summary Spectral Plots")
# plt.xlabel("Wavelength")
# plt.ylabel("Reflectance")
# plt.legend(labels = ['10nm Reflectance','1nm Reflectance','Derivative Spectra'])

```
